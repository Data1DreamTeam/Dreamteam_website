## Exercise 4: Shape the data
- Parse the `author_url` to create new column named `author`
- Sort the data by putting higher rates go first. If there are overlapping rates, try to put earlier years go first.
- **(Stretch Goal)** Examine how many books were published at each year and find lowest, highest rate of each year. 

```python
#--------------------------------- ADDING THE AUTHOR ---------------------------------------------

# 1) to add the author first we need to add the author column: df['author']

# 2) then we have to divide the cases where the name is actually in the url 
# from the ones where the name is missing: see if / else at the end of the line

# 3) we have to read through the url till the last / character and split the string there:
# x.split('/')[-1] allows this. 

# 4) we have to exlude what's before the . character: .split('.')[-1]

df['author'] = df['author_url'].apply(lambda x: x.split('/')[-1].split('.')[-1].replace('_', ' ') if pd.notnull(x) else "uknown")

print("------------------------------------")
print("Here are the first five authors")
print("------------------------------------")
df[['author_url', 'author']].head()
#-------------------------------------------------------------------------------------------------
```

```python
#--------------------------------------- SORTING ------------------------------------------
# Sort the data by 'rating' (ascending = false ) and then by 'year' (ascending = true) with
# the .sort_values funcion
df_sorted = df.sort_values(by=['rating', 'year'], ascending=[False, True])

#visual representation of the sorting
print("-------------------------------")
print("Here's the sorted dataset: ")
print("-------------------------------")
df_sorted[['rating','year']]

#-------------------------------------------------------------------------------------------
```

```python
#------------------------------ HIGHEST / LOWEST RATE OF THE YEAR -------------------------------

# Grouping the data by 'year' and aggregating to get the number of books, min and max rating
# .agg() used to apply multiple aggregating functions

yearly_stats = df.groupby('year').agg(
    books_published=('name', 'count'),     # Count how many books were published each year
    lowest_rating=('rating', 'min'),       # Find the lowest rating for each year
    highest_rating=('rating', 'max')       # Find the highest rating for each year
).reset_index()

yearly_stats

#---------------------------------------------------------------------------------------------
```

## Exercise 6: Investigate the relationship between the number of reviews and the average rating for books in the dataset cleaned-goodreads.csv procided.

- Calculate the correlation coefficient. Give me a short definition of this coefficient
- Create a scatter plot showing the relationship between these two features.
- Based on the plot and the correlation, provide a brief interpretation of the relationship.

### Python Tools: Use pandas and numpy for correlation, and matplotlib or seaborn for the scatter plot.

```python
# --------------------------------------- CORRELATION COEFFICIENT ----------------------------------------
# The correlation coefficient gives us an indication on how much two variables are linearly correlated 
# one to the other. It is defined as the covariance of the two variables divided by the product of their standard deviations.
# In formula:
# R(x,y) = ( COV (x,y) / sd(x)*sd(y) )

#importing numpy:
import numpy as np

#reading the file and calculating the correlation coefficient with .corrcoef 
df_2 = pd.read_csv("/Users/pietro/Documents/UNI/MADE/Data1/lab-02/data/cleaned-goodreads.csv")
corr_coeff = np.corrcoef(df_2['review_count'], df_2['rating'])[0, 1]

#printing
print("-------------------------------------")
print("The correlation coefficient is: ")
print(corr_coeff)
print("-------------------------------------")
# --------------------------------------------------------------------------------------------------------




# ---------------------------------- SCATTER PLOT ----------------------------------------------
#using matlpotlib for plotting
import matplotlib.pyplot as plt

#adjusting the dimensions of the image shown
plt.figure(figsize=(8,6))

#PLOTTING with .scatter function. Alpha sets the transparency of the dots
plt.scatter(df_2['review_count'], df_2['rating'], alpha=0.3)

#VISUALS of the plot: title, lables of the axis and adding a grid
plt.title('Scatter Plot: Review Count vs Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.grid(True)
#-----------------------------------------------------------------------------------------------

```

INTERPRETATION OF THE FIGURE
From the scatter plot we can see that the higher the number of reviews, the values of rating tend to get more stable in a range between 3.5 and 4.5. For low number of reviews the values of rating stand in a much wider range. 
This is confirmed by the bell shape that appears on the plot. 
The correlation coefficient ( -0.0378...) shows that the relationship between these variables is not linear since its value is very close to zero. This aspect is visually confirmed by the scatter plot where we can see that the values are not aligned. Instead they form a dense and wide area.

## Exercise 7: Calculate the following descriptive statistics for the numerical features (e.g., number of reviews, average rating, etc.):
- Mean
- Median
- Standard Deviation
- Range
- Create a histogram or box plot for at least one of the numerical features, highlighting any skewness or outliers.
    
### Python Tools: Use pandas for data manipulation and matplotlib or seaborn for visualization.

```python
#----------------------------------- CALCULATING DESCRIPTIVE STATISTICS -------------------------------

# Select the numerical columns to calculate statistics
num_col = ['review_count', 'rating', 'rating_count','year']

# Calculate MEAN, MEDIAN, STANDAR DEVIATION, MAX and MIN for the numerical features
# values are saved in statistics variable
# .agg() is used to apply multiple aggregating functions (mean,median...)
statistics = df_2[num_col].agg(['mean', 'median', 'std', 'min', 'max'])

# Calculate the RANGE as a difference between max and min
statistics.loc['range'] = statistics.loc['max'] - statistics.loc['min']

# Display the calculated descriptive statistics
print("-----------------------------------------------------------")
print("DESCRIPTIVE STATISTICS:")
print("-----------------------------------------------------------")
print(statistics)
print("-----------------------------------------------------------")
#--------------------------------------------------------------------------------------------------------
```

```python
#------------------------------------------ BOX PLOT ------------------------------------------
#importing seaborn
import seaborn as sns

# PLOTTING using a logarithmic scale to handle the skewness. .boxplot produces the boxplot of the review count
# log1p handles zero values safely,calculates the natural logarithm of 1 + x, which is work better than .log()
plt.figure(figsize=(8,6))
sns.boxplot(x=np.log1p(df_2['review_count']))  

# VISUALS of the plot
plt.title('Box Plot: Log of Review Count')
plt.xlabel('Log of Number of Reviews')
plt.show()
#-------------------------------------------------------------------------------------------------
```

<font color = 'mediumpurple'> <font size ='5'> 
For the descriptive statistics two of us were calculating each one independently but the best solution we agreed on was the one using the .agg() function. Like this everything can be calculated in one single step. 
Another important difference was that other codes were plotting a scatter plot. We opted for the boxplot because it really puts a focus on outliers (there are a lot of them if using this method). Another thing to mention is that the final code calculates .log1p() to manage numbers close to 0. Therefore the visual representation is more interesting with a boxplot.
